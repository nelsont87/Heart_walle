```python
from sklearn import tree
import pandas as pd
import numpy as np
import os
```

```python
dataframe = pd.read_csv("../Resources/heart.csv", sep=';')
dataframe
```

```python
random_forest_df = dataframe.drop(columns=['id'])
random_forest_df["age"] = (random_forest_df["age"]/365).apply(np.floor).astype(int) 
random_forest_df
```

```python
target = random_forest_df["cardio"]
target_names = ["negative", "positive"]
```

```python
data = random_forest_df.drop("cardio", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=1)
```

```python
print(X_train.shape, y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
rf.score(X_train, y_train)
```

```python
import pickle

Model_file = 'randomForest.pkl'

pickle.dump(rf, open(Model_file, 'wb'))
```
